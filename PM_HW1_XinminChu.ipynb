### Problem 1
Let $\mu \in \mathbb{R}$, $\sigma > 0$, and $k \in \mathbb{N}= \{ 0, 1, 2, \cdots \}$. Consider the integral
$$
I_k (\mu, \sigma) = \int_{- \infty}^{+ \infty} x^k e^{- \frac{(x - \mu)^2}{2 \sigma^2}} dx.
$$
In Lecture 1, we determined the case $k=0$ to be $I_0 (\mu, \sigma) = \sqrt{2 \pi \sigma^2}.$

- 1.1 Evaluate $I_1 (\mu, \sigma)$ in the case $k=1$.

From Lecture 1, we proved that, when $k=0$, $$
I_0 (\mu, \sigma) = \int_{- \infty}^{+ \infty} e^{- \frac{(x - \mu)^2}{2 \sigma^2}} dx = \sqrt{2 \pi \sigma^2}.
$$
When $k=1$, we have
$$
I_1 (\mu, \sigma) = \int_{- \infty}^{+ \infty} x e^{- \frac{(x - \mu)^2}{2 \sigma^2}} dx
$$
Let $x = \sigma X + \mu$, then $dx = \sigma dX$, and $X \ in (- \infty, + \infty)$.
$$
I_1 (\mu, \sigma) = \int_{- \infty}^{+ \infty} x e^{- \frac{(x - \mu)^2}{2 \sigma^2}} dx = \int_{- \infty}^{+ \infty} (\sigma X + \mu) e^{- \frac{X^2}{2}} \sigma dX \\
= \int_{- \infty}^{+ \infty} \sigma^2 X e^{- \frac{X^2}{2}} dX + \mu \int_{- \infty}^{+ \infty} e^{- \frac{X^2}{2}} \sigma dX
$$
Here, in the second term $\int_{- \infty}^{+ \infty} e^{- \frac{X^2}{2}} dX = \int_{- \infty}^{+ \infty} e^{- \frac{(x - \mu)^2}{2 \sigma^2}} dx = \sqrt{2 \pi \sigma^2}$ as we've shown in $I_0(\mu, \sigma)$.

Let $f(X) = X e^{- \frac{X^2}{2}}$, we see that $f(-X) = -X e^{- \frac{(-X)^2}{2}} = - f(X)$, so $f(X)$ is an odd function, and its integral over the symmetric interval from $- \infty$ to $+ \infty$ should be $0$.

Therefore, 
$$
I_1 (\mu, \sigma) = \sigma^2 \int_{- \infty}^{+ \infty} X e^{- \frac{X^2}{2}} dX + \mu \sqrt{2 \pi \sigma^2} = \boxed{\mu \sqrt{2 \pi \sigma^2}} = \mu I_0(\mu, \sigma)
$$

- 1.2 Evaluate $I_2 (\mu, \sigma)$ in the case $k=2$.

When $k=2$, we have
$$
I_2 (\mu, \sigma) = \int_{- \infty}^{+ \infty} x^2 e^{- \frac{(x - \mu)^2}{2 \sigma^2}} dx
$$
Let $x = \sigma X + \mu$, then $dx = \sigma dX$, as shown in part $1.1$.
$$
I_2 (\mu, \sigma) = \int_{- \infty}^{+ \infty} x^2 e^{- \frac{(x - \mu)^2}{2 \sigma^2}} dx = \int_{- \infty}^{+ \infty} (\sigma X + \mu)^2 e^{- \frac{X^2}{2}} \sigma dX \\
= \int_{- \infty}^{+ \infty} \sigma^2 X^2 e^{- \frac{X^2}{2}} \sigma dX + \int_{- \infty}^{+ \infty} 2\sigma X \mu e^{- \frac{X^2}{2}} \sigma dX + \int_{- \infty}^{+ \infty} \mu^2 e^{- \frac{X^2}{2}} \sigma dX \\
= \sigma^3 \int_{- \infty}^{+ \infty} X^2 e^{- \frac{X^2}{2}} dX + 2\sigma^2 \mu \int_{- \infty}^{+ \infty} X e^{- \frac{X^2}{2}} dX + \mu^2 \int_{- \infty}^{+ \infty} e^{- \frac{X^2}{2}} \sigma dX
$$

In the first term, $X^2 e^{- \frac{X^2}{2}} dX$ is an even function, so, $\int_{- \infty}^{+ \infty} X^2 e^{- \frac{X^2}{2}} dX = 2 \int_{0}^{+ \infty} X^2 e^{- \frac{X^2}{2}} dX$.

Since $d(e^{- \frac{X^2}{2}}) = e^{- \frac{X^2}{2}} (- \frac{X^2}{2})'dX = e^{- \frac{X^2}{2}} (-X) dX$, such that, $X e^{- \frac{X^2}{2}} dX = - d(e^{- \frac{X^2}{2}})$. We can rewrite it as following and use integral by part,
$$
\int_{0}^{+ \infty} X^2 e^{- \frac{X^2}{2}} dX = \int_{0}^{+ \infty} X (X e^{- \frac{X^2}{2}} dX) = \int_{0}^{+ \infty} X (- d(e^{- \frac{X^2}{2}})) \\
= - \int_{0}^{+ \infty} X de^{- \frac{X^2}{2}} = - ( X e^{- \frac{X^2}{2}} \mid_0 ^ {+ \infty} - \int_{0}^{+ \infty} e^{- \frac{X^2}{2}} dX ) \\
= - ( 0 - \int_{0}^{+ \infty} e^{- \frac{X^2}{2}} dX ) = \int_{0}^{+ \infty} e^{- \frac{X^2}{2}} dX = \dfrac{\sqrt{2 \pi}}{2}
$$

In the second term, as shown in part $1.1$ above, $\int_{- \infty}^{+ \infty} X e^{- \frac{X^2}{2}} dX = 0$.

In the third term, $\int_{- \infty}^{+ \infty} e^{- \frac{X^2}{2}} \sigma dX = \sqrt{2 \pi \sigma^2}$

All in together, we get,
$$ I_2(\mu, \sigma) = \sigma^3 \int_{- \infty}^{+ \infty} X^2 e^{- \frac{X^2}{2}} dX + 2\sigma^2 \mu \int_{- \infty}^{+ \infty} X e^{- \frac{X^2}{2}} dX + \mu^2 \int_{- \infty}^{+ \infty} e^{- \frac{X^2}{2}} \sigma dX \\
= \sigma^3 \times 2 \times \dfrac{\sqrt{2 \pi}}{2} + 2 \sigma^2 \mu \times 0 + \mu^2 \times \sqrt{2 \pi \sigma^2} \\
= \sigma^2 \sqrt{2 \pi \sigma^2} + \mu^2 \sqrt{2 \pi \sigma^2} \\
= \boxed{(\sigma^2 + \mu^2) \sqrt{2 \pi \sigma^2}} = (\sigma^2 + \mu^2)I_0(\mu, \sigma)
$$

### Problem 2

In lecture2, we defined probability models $P$ and saw that if an experiment has a finite sample space $S$, there is a distinguished probability model $P_{unif}$ for the experiment which posits that all sample outcomes $s \in S$ are equally likely. This $P_{unif}$ is the *uniform probability model*. In this model, we calculate likelihoods of events by *Cardano's rule*
$$
P_{unif}(A) = \frac{number \, of \, outcomes \, in \, A}{number \, of \, outcomes \, in \, S}
$$
In Lecture2, we considered the experiment of tossing 4 coins, each with outcome $H$ or $T$, and described a finite sample sapce $S$ for this experiment with $2^4 = 16$ sample outcomes such as $s = HHTH$. Let $B_i$ be the event that the *i*th coin is $H$, e.g. $s = HHTH \notin B_3$.

Here, the sample space $S$ is:
$$
S = \left\{ \begin{array} HHHH, \\HHHT, HHTH, HTHH, THHH,\\ HHTT, HTHT, THHT, THTH, HTTH, TTHH, \\TTTH, TTHT, THTT, HTTT, \\TTTT \end{array} \right\}
$$
and then, by definition of $B_i$, $i \in \{ 1, 2, 3, 4 \}$
$$
B_1 = \{ HHHH, HHHT, HHTH, HTHH, HHTT, HTHT, HTTH, HTTT \};\\
B_2 = \{ HHHH, HHHT, HHTH, THHH, HHTT, THHT, THTH, THTT \}; \\
B_3 = \{ HHHH, HHHT, HTHH, THHH, HTHT, THHT, TTHH, TTHT \}; \\
B_4 = \{ HHHH, HHTH, HTHH, THHH, THTH, HTTH, TTHH, TTTH \}.
$$

- 2.1 Are the events $B_2$ and $B_4$ mutually exclusive?

Answer: $\fbox{No. They are not mutually exclusive}$.

$B_2 \cap B_4 = \{ HHHH, HHTH, THTH, THHH \} \neq \phi$, so $B_2$ and $B_4$ are not mutually exclusive.

- 2.2 Calculate $P(B_2 \cap B_4)$. *Note that $B_2B_4 = B_2 \cap B_4$ in Ross*

$$
P_{unif}(B_2 \cap B_4) = \frac{number \, of \, outcomes \, in \, (B_2 \cap B_4)}{number \, of \, outcomes \, in \, S} = \frac{4}{16} = \boxed{\frac{1}{4}} = 0.25 = \boxed{25 \%}
$$

- 2.3 Calculate $P(B_2)P(B_4)$.

$P(B_2)P(B_4) = \dfrac{number \, of \, outcomes \, in \, B_2}{number \, of \, outcomes \, in \, S} \times \dfrac{number \, of \, outcomes \, in \, B_4}{number \, of \, outcomes \, in \, S} =\dfrac{8}{16} \times \dfrac{8}{16} = \boxed{\dfrac{1}{4}}$

- 2.4 Are the events $B_2$ and $B_4$ independent? Why or why not?

Answer: $\fbox{Yes. They are independent.}$

Since $P(B_2 \cap B_4) = \dfrac{1}{4} = P(B_2)P(B_4)$, the events $B_2$ and $B_4$ are independent.

**Definition:** Suppose $A$, $B$ are events in a sample space, $S$, of a probability model $P$, we say that $A$ and $B$ are *indepentdent* events according to $P$ if $P(A \cap B) = P(A)P(B).$

Event $C(two \, rolls \, sum \, to \, 8) = \left\{ \{2,6 \}, \{3,5 \}, \{4,4 \}, \{5,3 \}, \{6,2 \},\right\}$, and event $C \cap A = \phi$. So, $P(C \cap A) = \boxed{0}$.

### Problem 4

A set $S$ of voters participated in a recent election between Alice and Bob. In total, $3355$ voters participated, $1588$ voted for Bob, $725$ were first-time voters, and $465$ were first-time voters who voted for Bob.

**Suppose each voter must and only vote for one person from Alice and Bob, no other name and no empty.** Let $A$ be the event that Alice was voted, $B$ be the event that Bob was voted, and $F$ be the event that the person is the first-time voter. 

From the known conditions and according to uniform probability model $P$, we konw,
1. $P(B) = \dfrac{1588}{3355}$, so $P(A) = \dfrac{3355 - 1588}{3355} = \dfrac{1767}{3355}$. 
1. $P(F) = \dfrac{725}{3355}$, so $P(\neg F) = 1 - P(F) = \dfrac{2630}{3355}$.
1. $P(B \cap F) = \dfrac{465}{3355}$, so $P(A \cap F) = P(F) - P(B \cap F) = \dfrac{725 - 465}{3355} = \dfrac{260}{3355}$.

- 4.1 Suppose you pick a voter uniformly at random. What is the probability that this voter voted for Bob?

$P(B) = \dfrac{1588}{3355} = \boxed{0.4733} = \boxed{47.33 \%}$

- 4.2 Suppose you pick a voter uniformly at random from the set of voters who voted for Alice. What is the probability that this voter was a first-time voter?

$P(F|A) = \dfrac{P(F \cap A)}{P(A)} = \dfrac{260 / 3355}{1767 / 3355} = \boxed{0.1471} = \boxed{14.71 \%}$

- 4.3 Suppose the votes for the election are shuffled and stacked in a pile. What is the probability that the vote at the top of the pile is for Alice?

Sample space for one vote is $S = \{Alice, Bob \}$. Under my assumption (Suppose each voter must and only vote for one person from Alice and Bob, no other name and no empty) before, $P(the \, top \, of \, the \, pile \, is \, Alice) = \boxed{\dfrac{1}{2}} = \boxed{50 \%}$

- 4.4 Suppose the votes for the election are shuffled and stacked in a pile, but then this time the vote at the top of the pile is removed. What is the probability that the *second* vote is for Alice?

In this condition, each vote is no relation to others (independent), and removing the top vote of the pile doesn't affect the content on the second vote in that pile. So, $P(the \, second \, of \, the \, pile \, is \, Alice \, after \, removing \, the \, top \, vote) = \boxed{\dfrac{1}{2}} = \boxed{50 \%}$

- 4.5 Who won the election, Alice or Bob?

If the winning rule is simple most, then comparing the number of votes for Alice and Bob, the winner is who got more votes in this election. Since $1588$ voted for Bob, then $3355 - 1588 = 1767$ voted for Alice, here Alice got more votes ($1767 > 1588$), so, finally $\fbox{Alice}$ won the election.

### Problem 5

A computer manufacturer uses chips from three suppliers. Based on past performance, she knows that the chips from supplier $A$ will fail within 1 year with probability $0.01$, the chips from suppliers $B$ will fail within 1 year with probability $0.02$, and the chips from supplier $C$ will fail within 1 year with probability $0.05$. She buys $50 \%$ of her chips from supplier $A$, $40 \%$ from supplier $B$, and $10 \%$ from supplier $C$.

Let $F$ be the event that a chip failed within 1 year, $A, B, C$ are events that a chip comes from supplier $A, B$ or $C$, respectively. From the known conditions, according to uniform probability model $P$, we have,
$$
P(A) = 0.5, P(B) = 0.4, P(C) = 0.1; \\
P(F|A) = 0.01, P(F|B) = 0.02, P(F|A) = 0.05
$$

- 5.1 Suppose a chip is chosen uniformly at random out of one of her computers on the belt in the factory. What is the probability it fails within 1 year?

Supplier $A$, supplier $B$, and supplier $C$ are different, without any joint part, that is, $A$, $B$, and $C$ are mutually exclusive (disjoint). In math form, $A \cap B = \phi$, $B \cap C = \phi$, and $C \cap A = \phi$. Meanwhile, $P(A) + P(B) + P(C) = 0.5 + 0.4 + 0.1 = 1$, we can say, $A \cup B \cup C = S$. So, event $A$, $B$, and $C$ will partition the sample space perfectly.

$$
P(F) = P(F|A)P(A) + P(F|B)P(B) + P(F|C)P(C) = 0.01 \times 0.5 + 0.02 \times 0.4 + 0.05 \times 0.1 = \boxed{0.018}.
$$

- 5.2 Suppose a chip in one of her computers has failed within 1 year. What is the probability that it came from supplier $B$?

$$
P(B|F) = \dfrac{P(F|B)P(B)}{P(F|A)P(A) + P(F|B)P(B) + P(F|C)P(C)} = \dfrac{P(F|B)P(B)}{P(F)} = \dfrac{0.04 \times 0.4}{0.018} = \boxed{0.8889}
$$
